```python
#Minimum, Maximun, Median, Q1, Q2, Q3, IQR
```

```python
import numpy as np
```

```python
lst_marks = [45,32,56,75,89,54,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximum = map(int, np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0]))
```

```python
print(minimum,Q1,median,Q3,maximum)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
```

```python
print(lower_fence)
print(higher_fence)

#Any values in the list less than 1.5 or greater than 141.5 is considered as an outlier
```
